```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=load_iris()
X=data.data
y=data.target

print(data.feature_names)
print(data.target_names)
```

```python
X[y==0,2]
```

```python
plt.figure(figsize=(8,5))
plt.plot(X[y==0,2],X[y==0,3],'b.',label='class1')
plt.plot(X[y==1,2],X[y==1,3],'g*',label='class2')
plt.plot(X[y==2,2],X[y==2,3],'y^',label='class2')
plt.xlabel('Petal Length')
plt.xlabel('Petal width')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
X=X[:,2:]
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X,y)
y_pred=model.predict(X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y,y_pred)
```

```python
model.n_neighbors
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
plt.figure(figsize=(8,5))
disp = DecisionBoundaryDisplay.from_estimator(model,X,response_method='predict')
plt.plot(X[y==0,0],X[y==0,1],'b.',label='class1')
plt.plot(X[y==1,0],X[y==1,1],'g*',label='class2')
plt.plot(X[y==2,0],X[y==2,1],'y^',label='class2')

plt.show()
```

## Elbow technique


```python
range(2,10,1)
```

```python
accuracy_score_list=[]
for k in range(2,10,1):
  model=KNeighborsClassifier(n_neighbors=k)
  model.fit(X,y)
  y_pred = model.predict(X)
  accuracy_score_list.append(accuracy_score(y,y_pred))

plt.plot([2,3,4,5,6,7,8,9],accuracy_score_list)
plt.grid()
plt.show()
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
y_pred = model.predict(X)
```

```python
accuracy_score(y,y_pred)
```

 The 'metric' parameter of KNeighborsClassifier must be a str among {'correlation', 'cosine', 'dice', 'braycurtis', 'sokalmichener', 'p', 'chebyshev', 'infinity', 'matching', 'l2', 'kulsinski', 'minkowski', 'precomputed', 'hamming', 'nan_euclidean', 'russellrao', 'wminkowski', 'sokalsneath', 'manhattan', 'l1', 'jaccard', 'haversine', 'cityblock', 'pyfunc', 'sqeuclidean', 'canberra', 'euclidean', 'seuclidean', 'mahalanobis', 'rogerstanimoto', 'yule'}

```python
model=KNeighborsClassifier(n_neighbors=3,metric='euclidean')
model.fit(X,y)
y_pred = model.predict(X)
plt.figure(figsize=(8,5))
disp = DecisionBoundaryDisplay.from_estimator(model,X,response_method='predict')
plt.plot(X[y==0,0],X[y==0,1],'b.',label='class1')
plt.plot(X[y==1,0],X[y==1,1],'g*',label='class2')
plt.plot(X[y==2,0],X[y==2,1],'y^',label='class2')

plt.show()
```
